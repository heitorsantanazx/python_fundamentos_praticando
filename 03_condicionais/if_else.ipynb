Estudos de Condicionais

```python
saldo = 400
valor_saque = int(input('Informe o valor do saque: '))
idade = int(input('Informe a sua idade: '))

if saldo >= valor_saque:
    saldo -= valor_saque
    print('Saque feito com sucesso')

elif idade >= 25 and valor_saque > 500:
    novo_valor_saque = valor_saque - saldo
    print('Você pega de empréstimo R$ ', novo_valor_saque)

    novo_saldo = - novo_valor_saque
    print('Seu novo saldo é: ', novo_saldo)

else:
    print('Saque insuficiente e sem possibilidade de empréstimo.')
```

```python
idade_usuario = 25
saldo = 400
valor_saque = 700

if saldo >= valor_saque :
   saldo -= valor_saque
   # comandos para liberar o dinheiro
   print('Saque efetuado com sucesso!')
elif saldo + 500.0 >= valor_saque and idade_usuario > 25 : # teste para se saber se o cheque especial é suficiente e a idade é suficiente
   saldo -= valor_saque
   # comandos para liberar o dinheiro
   print('Saque efetuado com sucesso com uso do cheque especial! Cuidado!')
else :
   print('Saque não efetuado: Saldo insuficiente!')
```

```python
saldo = 500
idade_usuario = 21
cheque_especial = 500
valor_saque = 300

# teste para saber se possui limite para o saque,
# caso positivo, o saque é efetuado
if saldo >= valor_saque:
   saldo -= valor_saque
   print('Saque efetuado com sucesso!')
   print('Saldo: R$%0.2f' % saldo)

# mesmo sem saldo, usuários com mais de 25 anos de idade,
# podem sacar até 500 de cheque especial
elif idade_usuario >= 25 and saldo + cheque_especial >= valor_saque:
   saldo -= valor_saque
   print('Saque efetuado com sucesso!')
   print('Cheque especial: R$%0.2f' % saldo)

# cenário onde o dinheiro não é liberado
else:
   print('Saque não efetuado!')
   print('Saldo insuficiente!')
```
