```python
import math

terreno = [
    [42.7, 71.9], [38.9, 80.3], [56.3, 88.9], [70.7, 83.9],
    [74.5, 69.7], [56.8, 56.5], [47.9, 60.4], [36.8, 62.6]
]

arame_por_segmento = []
estacas_por_segmento = []
comprimento_total_arame = 0
total_estacas = 0

for i in range(len(terreno)):
    p1 = terreno[i]
    # O último segmento conecta o último ponto com o primeiro
    p2 = terreno[(i + 1) % len(terreno)] 

    # Calculando o comprimento do segmento usando o Teorema de Pitágoras
    comprimento_segmento = math.sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2)

    # Adicionando o comprimento ao arame por segmento
    arame_por_segmento.append(round(comprimento_segmento, 2))
    
    # Calculando o número de estacas (sem contar os extremos)
    # A regra é: número de estacas = parte inteira de (comprimento / 2)
    num_estacas_segmento = math.floor(comprimento_segmento / 2)
    estacas_por_segmento.append(num_estacas_segmento)

    # Calculando o comprimento total do arame
    comprimento_total_arame += comprimento_segmento
    
    # A estaca é colocada em cada ponto, então o total é o número de pontos
    total_estacas += (num_estacas_segmento + 1)

print(f"Arame por segmento: {arame_por_segmento}")
print(f"Estacas por segmento (sem extremos): {estacas_por_segmento}")
print(f"Comprimento total de arame: {round(comprimento_total_arame, 2)} metros")
print(f"Quantidade total de estacas: {total_estacas}")
```
