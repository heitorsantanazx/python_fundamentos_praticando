# Calculadora v1

```python
print('Bem vindo à calculadora \nEscolha uma operação:Soma, Subtração, Multiplicação ou Divisão')
opcao = input('Digite a operação: ')

if opcao == 'Soma':
    print('Você escolheu soma, vamos ao cálculo!')
    primeiroNumero = float(input('Digite o primeiro número: '))
    segundoNumero = float(input('Digite o segundo número: '))

    soma = primeiroNumero + segundoNumero

    print(f'O resultado da soma foi {soma}')

elif opcao == 'Subtração':
    print('Você escolheu subtração, vamos ao cálculo!')
    primeiroNumero = float(input('Digite o primeiro número: '))
    segundoNumero = float(input('Digite o segundo número: '))

    subt = primeiroNumero - segundoNumero

    print(f'O resultado da soma foi {subt}')

elif opcao == 'Multiplicação':
    print('Você escolheu multiplicação, vamos ao cálculo!')
    primeiroNumero = float(input('Digite o primeiro número: '))
    segundoNumero = float(input('Digite o segundo número: '))

    multi = primeiroNumero * segundoNumero

    print(f'O resultado da soma foi {multi}')

elif opcao == 'Divisão':
    print('Você escolheu divisão, vamos ao cálculo!')
    primeiroNumero = float(input('Digite o primeiro número: '))
    segundoNumero = float(input('Digite o segundo número: '))

    divisao = primeiroNumero / segundoNumero

    print(f'O resultado da soma foi {divisao}')

else:
    print('Operação inválida! Tente novamente')
```
