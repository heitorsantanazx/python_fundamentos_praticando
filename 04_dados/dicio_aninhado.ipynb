Treinando com dicionário

```python
familia = {}

familia['pai'] = {'nome': 'João', 'idade': 40, 'passatempo': 'futebol'}
familia['mãe'] = {'nome': 'Maria', 'idade': 40, 'profissão': 'professora'}
familia['filho 01'] = {'nome': 'Pedro', 'idade': 20, 'salário': 1000.0}
familia['filho 02'] = {'nome': 'João', 'idade': 17}
```

```python
for papel, dados in familia.items():
    print('Dados do(a) {}:' .format(papel))
    for chave, dado in dados.items():
        print('{} -> {}' .format(chave, dado))
    print('')
```
