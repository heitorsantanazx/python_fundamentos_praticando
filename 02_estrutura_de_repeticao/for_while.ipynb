# Estudos de FOR e WHILE

```python
for contador in range(5):
    print('Isto é um laço.')
```

```python
for contador in range(3,5):
    contador += 0
    print(contador)
```

```python
vezes = int(input('Digite um número: '))

for repeticao in range(vezes):
    print('Outro exemplo de loop!')
```

```python
texto = 'computador'
numero = 0

for contador in texto: #neste caso vai repetir de acordo com o número de caracteres em texto que são 10
    numero += 1
    print('Python versão', numero)
```

```python
texto = 'computador'

for contador in texto:
    print(contador) #vai imprimir o dado na variável texto na vertical
```

```python
texto = 'computador de heitor'

for contador in texto:
    print(contador, end='') #vai imprimir na horizontal
```

```python
a = 0
numero = 0

while a < 4:
    a += 1
    print('Python versão', a)
    if a == 4:
        print('Esta é a versão final!')
```

```python
a = 0
numero = 0

while a < 10:
    a += 1
    print('Python versão', a)
    if a == 4:
        print('Esta é a versão final!')
        break #serve para pausar
```

```python
a = 0

while a < 10:
    a += 1
    if a == 5:
        pass #encontra o 5, mas imprime e continua
    print('Python versão', a)
```

```python
a = 0

while a < 10:
    a += 1
    if a == 5:
        continue #ele encontra 5, mas não imprime e continua o programa
    print('Python versão', a)
```
