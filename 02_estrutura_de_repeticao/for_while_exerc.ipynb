Exercícios com For e While

```python
x = 0

for x in range(2, 100):
    if x % 2 != 0:
        print(x)
```

```python
num_str = input('Digite um número qualquer: ')

while int(num_str) % 2 == 0: #enquanto número for par continua rodando o loop
    print('Você digitou um número par! Vou pedir que digite outro.')

    num_str = input('Digite outro número qualquer: ')

print('Parabéns, você digitou um número ímpar! Agora o programa vai seguir...')
```

```python
foram = 0
gostaram = 0

n = int(input('Digite o número de pessoas na sala: '))

for i in range(n):
  foi = input('Digite S caso a ' + str(i + 1) + '-a pessoa tenha ido ao Novo Shopping e N caso negativo: ')
  if foi == 'N':
     continue
     foram += 1
  gostou = input('Digite S caso a vista tenha sido boa e N caso contrário: ')
  if gostou == 'S':
     gostaram += 1

print('De um total de ' + str(n) + ' pessoas, ' + str(foram) + ' visitaram o Novo Shopping e deste total, ' + str(gostaram) + ' gostaram da visita.')
```
